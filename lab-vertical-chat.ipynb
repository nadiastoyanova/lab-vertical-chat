```python
#if you need an API Key from OpenAI
#https://platform.openai.com/account/api-keys

from openai import OpenAI
import os
from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv())

OPENAI_API_KEY  = "test-key" or os.getenv('OPENAI_API_KEY')
```

```python
client = OpenAI(
    # This is the default and can be omitted
    api_key=OPENAI_API_KEY,
)

def continue_conversation(messages, temperature=0):
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=messages,
        temperature=temperature,
    )
    #print(str(response.choices[0].message["content"]))
    return response.choices[0].message.content
```

```python
def add_prompts_conversation(_):
    #Get the value introduced by the user
    prompt = client_prompt.value_input
    client_prompt.value = ''

    #Append to the context the User prompt.
    context.append({'role':'user', 'content':f"{prompt}"})

    #Get the response.
    response = continue_conversation(context)

    #Add the response to the context.
    context.append({'role':'assistant', 'content':f"{response}"})

    #Update the panels to show the conversation.
    panels.append(
        pn.Row('User:', pn.pane.Markdown(prompt, width=600)))
    panels.append(
        pn.Row('Assistant:', pn.pane.Markdown(response, width=600)))

    return pn.Column(*panels)
```

This study evaluated several variations of structured “OrderBot” prompts designed to guide GPT in a food-ordering workflow. The primary objective was to assess how reliably the model follows instructions, avoids hallucinations, and maintains a consistent conversational structure when constrained by a predefined menu and interaction flow.

Across most prompt variations, GPT performed well when the instructions were explicit and sequential. Prompts that clearly defined the steps—greeting the customer, collecting the full order (including drinks), clarifying options, summarizing, confirming, and finally collecting payment—resulted in accurate and predictable behavior. The presence of a structured menu with clearly listed items, prices, and customization options significantly improved performance by anchoring the model’s responses to a fixed set of choices.

However, some variations did not work as effectively. When prompts lacked strict constraints (for example, failing to explicitly state “only use items from the menu”), GPT occasionally hallucinated by suggesting items, sizes, or combinations that were not listed. In a few cases, the model inferred “standard” options (such as default sizes or included toppings) without asking for clarification, which led to incomplete or incorrect orders. Another issue arose with poorly formatted or dense menus—if items, prices, and categories were not clearly separated, the model sometimes misinterpreted pricing tiers or skipped necessary follow-up questions.

Additionally, prompts that were vague about tone or response length resulted in inconsistent conversational style. Some outputs became overly verbose, while others skipped important steps like final confirmation before payment. Variations that omitted reminders to “wait for the full order” led GPT to prematurely summarize or finalize orders, disrupting the intended flow.

Overall, the main failure modes were tied to ambiguity, lack of constraints, and inconsistent formatting. When instructions were incomplete or loosely defined, GPT filled in gaps with assumptions, increasing the likelihood of hallucination or logical errors.

The most important takeaway is that prompt clarity and structure directly determine output quality. Explicit, step-by-step instructions significantly improve reliability, especially in transactional tasks like order-taking. Clearly stating constraints—such as restricting responses strictly to the provided menu—helps prevent hallucinations.

I also learned that formatting matters as much as content. Well-organized menus with consistent categories, pricing, and option labels make it easier for the model to interpret and respond accurately. Poor formatting introduces confusion and increases the chance of mistakes.

Another key insight is the value of redundancy. Repeating critical rules (e.g., always confirm the full order before proceeding) reinforces correct behavior. However, this must be balanced with brevity to avoid overwhelming the model with unnecessary detail.

Finally, I learned that even small prompt adjustments—such as specifying tone, response length, or clarification requirements—can have a noticeable impact on output consistency. Effective prompt design is iterative and benefits from testing multiple variations to identify weaknesses.

In conclusion, GPT can perform structured tasks very effectively when given precise, well-formatted, and constrained instructions. Most errors are not random but stem from gaps in prompt design, highlighting the importance of clarity, specificity, and thoughtful structuring.
